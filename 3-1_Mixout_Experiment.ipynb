```python
import os
import zipfile
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import cv2
```

```python
os.chdir("..//")
```

```python
zipData = zipfile.ZipFile("SEG_Train_Mask_Images_Dataset_20220509_Width_Height.zip")
original_image = [i for i in zipData.namelist() if "Mask" not in i and ".jpg" in i][:2]
mask_image = [i for i in zipData.namelist() if "Mask" in i and ".jpg" in i][:2]
```

```python
original_image1 = plt.imread(zipData.open(original_image[0]))
mask_image1 = plt.imread(zipData.open(mask_image[0]))
original_image2 = plt.imread(zipData.open(original_image[1]))
mask_image2 = plt.imread(zipData.open(mask_image[1]))
```

```python
plt.figure(figsize = (10, 10))
plt.subplot(2, 2, 1)
plt.imshow(original_image1)
plt.axis("off")
plt.subplot(2, 2, 2)
plt.imshow(original_image2)
plt.axis("off")
plt.subplot(2, 2, 3)
plt.imshow(mask_image1, cmap = "gray")
plt.axis("off")
plt.subplot(2, 2, 4)
plt.imshow(mask_image2, cmap = "gray")
plt.axis("off")
plt.show()
```

```python
lmbda = 0.5
mixout = lmbda * original_image2 / 255 + (1-lmbda) * original_image1 / 255
plt.imshow(mixout)
plt.show()
```

```python
mixout_mask = lmbda * mask_image2 / 255 + (1-lmbda) * mask_image1 / 255
plt.imshow(mixout_mask, cmap = "gray")
plt.show()
```
